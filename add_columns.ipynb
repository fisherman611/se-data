```python
import pandas as pd

progress_df = pd.read_csv('data/progress.csv')
result = progress_df.groupby('u_id')['score'].sum().reset_index()
points = result['score'].to_list()
```

```python
import random

df['points'] = points
# List of avatar paths
avatar_paths = [
    '/images/default_avatar.png',
    '/images/avatar1.png', '/images/avatar2.png', '/images/avatar3.png',
    '/images/avatar4.png', '/images/avatar5.png', '/images/avatar6.png',
    '/images/avatar7.png'
]

# Add 'avatar' column with random choice from avatar_paths
df['avatar'] = [random.choice(avatar_paths) for _ in range(len(df))]

df
```
